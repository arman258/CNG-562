```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
pd.set_option('display.max_columns', 500)
cancer = load_breast_cancer()
print (cancer.keys())
df_cancer = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df_cancer['target'] = pd.Series(cancer.target)
#print (pd.isnull(df_cancer).any())

df_cancer.columns=df_cancer.columns.str.replace(" ","_")
df_cancer.shape
```

```python
X = df_cancer.drop('target', axis=1)
y = df_cancer['target']

print (X.shape)
print (y.shape)
```

```python
### taking into account Mean

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

print (X_train.shape)

print (y_train.shape)

```

# __Logistic Regression__

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score
clf = LogisticRegression(random_state = 0, solver='liblinear')
clf.fit(X_train, y_train)

print ("train accuracy is :",format(clf.score(X_train, y_train)))

clf_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
print (confusion_matrix(y_test , clf_pred))


print(accuracy_score(y_test,clf_pred).round(3))
print(precision_score(y_test,clf_pred).round(3))

accuracy_score(y_train, y_train.predict)



```

# __KNN__ 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score
import numpy as np

k_range = list(range(1,50))
error = []

for k in k_range :
    
    neigh = KNeighborsClassifier(n_neighbors = k)
    neigh.fit(X_train, y_train)
    neigh_pred = neigh.predict(X_test)
    score = accuracy_score(y_test,neigh_pred)
    error.append(score)
        
print ("train accuracy is :",format(neigh.score(X_train, y_train)))

plt.plot (k_range,error)
print ("mean best score w/o kfold :",format(np.mean(error)))
print ("max best score w/o kfold :",format(np.max(error)))

best_k = k_range[error.index(np.max(error))]

print ("best k w/o kfold :",format(best_k))

#from sklearn.metrics import confusion_matrix
#print (confusion_matrix(y_test , neigh_pred))
```

## __KNN Cross Valiadation__ 

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import cross_val_score

#del cv_score

cv_score = []
k_range = list(range(1,50))

for k in k_range :
    
    knn = KNeighborsClassifier(n_neighbors = k)
    score_cv = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    cv_score.append(np.mean(score_cv))
    
print ("mean cv score w/ 10 kfold :",format(np.mean(cv_score)))
print ("max cv score w/ 10 kfold :",format(np.max(cv_score)))

best_k_fold = k_range[cv_score.index(np.max(cv_score))]
print ("best k w/ kfold :",format(best_k_fold))
```
